# Teoría Computacional sobre Transformación de Empresa

Queremos observar la influencia de un modelo sobre el dirrecionamiento de una empresa, queremos ver el uso de herramientas de computación que nos puedan ayudar a tomar decisiones informadas.

Recordamos que en la formulación matemática existen los siguiente conceptos claves que ayudan a describir la situación de la empresa, estos parámetros son: $\alpha$ que defiine el grado de volatilidad del mercado, $\beta$ que especifica el grado de incercia de la empresa, $\delta$ que cuantifica el aprendizaje de la empresa respecto al mercado, y $\lambda$ que especifica las actitudes de los clientes hacia la oferta de mercado.

Recordamos que el valor del mercado (VM) es definido como una función de la diferencia entre los deseos del mercado y la oferte $XO(t)$ Modelada por:

$$
    VM(t) = e^{ - \lambda | XM(t) - XO(t) | }
$$
Donde $XM(t)$ es la preferncia del mercado que es dependiente de un proceso estocástico.

Tambien tenemos el valor percibido de mercado por parte de la empresa esta dado por:

$$
    VP(t+1) = \delta(t-1)M(t) + \delta(t-2)M(t-1) + \cdots
$$

Recordamos que nuestro objetivo es conseguirr $XO$ lo más cerca posible de $XM$ para producir la más alta $VM$ 


Tenemos los siguientes párametros $\alpha = 0.9$, $\beta = 0.8$, $\delta = 5 $

Vamos a hacer las gráficas del valor del mercado percibido (VP) contra el tiempo (150 periodos), variando cada uno de los parámetros.

$$ VP(t + 1) = \delta(t-1) VM(t) + \delta_2 VM( t-1 ) + \cdots $$

```julia
function PreferenciasMercado(t::Number,α::Number)
    if t > 100
        d2 = Normal(t-1, √10)
        sol = (100.0*α + rand(d2, 1) *(1-α))
    end
    if t == 0
        sol = 100.0
    end
    if t <= 100
        sol = (100.0*α + rand(d1, 1) *(1-α))
    end
    return sol[1]
end
```

```julia
function PreferenciasFuturas(t::Number, α::Number)
    return 0.4*PreferenciasMercado(t-1,α) + 0.3*PreferenciasMercado(t-2, α) + 0.3*PreferenciasMercado(t-3, α)
end
```

```julia
function OfertadelaCompañia(t, β, α)
    if t == 0
        return 100.0
    end
    if t == 1
        return 100.0
    end
    if t > 1
        return(β*OfertadelaCompañia(t-1,β, α) + PreferenciasFuturas(t + 3, α)*(1-β))
    end
end
```

```julia
function ValorMercado(t, α, β, λ)
    if !(t == 0)
        exp(- λ * abs(PreferenciasMercado(t, α) - OfertadelaCompañia(t, β, α)))
    elseif (t < 1)
        return 0
    else
        return 1
    end
end
```

```julia
function delta(t)
    if t == 1
        return 1
    else
        return 0
    end
end
```

```julia
function ValorPMercado(t, δ, α, β, λ)
    if t < 0
        return 1
    elseif t == 0
        return 1
    else
        return ValorMercado(t - δ, α, β, λ)
    end
end
            
```
